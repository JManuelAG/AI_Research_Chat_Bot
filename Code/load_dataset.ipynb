```python
import pandas as pd
import numpy as np
from datasets import Dataset, DatasetDict
from sklearn.model_selection import train_test_split
from huggingface_hub import notebook_login
```

The following code is used to create a dataset and push it to the Hugging Face Space 

```python
# Read a .dat document
file_path = "../Datasets/bert_dataset.csv"
dataset = pd.read_csv(file_path)

# Inspect the initial look of the data frame
dataset
```

```python
# Read a .dat document
file_path = "../Datasets/bert_test.csv"
test = pd.read_csv(file_path)

# Inspect the initial look of the data frame
test
```

Create a Dataset Dictionarty to Load to the Hub

```python
# Create the datasets for each one
dataset_dic = Dataset.from_dict({'question': dataset.question.values, 'answer': dataset.answer.values})
test_dic = Dataset.from_dict({'question': test.test_question.values, 'answer': test.original_question.values})

# Merge all to one dataset dictionary
dataset_full = DatasetDict({
    'dataset': dataset_dic,
    'test': test_dic
})

# View the final dataset
dataset_full
```
